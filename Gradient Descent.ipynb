```python
# Runs gradient descent method, to find the minimum value of a function, starting from an initial value for x and iterating using X1=X0-(learning_rate)*(dy/dx)
#To get the next value of x. The algorithm repeats until termination, when the precision value between x values is reached, or the max number of iterations is reached
#The function to solve is 3x^2-6x+4, where the minimum is at x=1
# The derivative is 6x-6, and given as the function df below


initial_x=5 #initial value for x
rate=0.01 # This is learning rate
precision=0.0001 #This is precision value, tells us when to stop algorithm
maximum_iterations=500 #max iterations to use

previous_step_size=1 #initialize step size
num_of_iters=0 #initialize number of iterations

df=lambda x: (6*x-6)

while previous_step_size>precision and num_of_iters<maximum_iterations:
    new_x = initial_x
    cur_x=new_x-rate*df(new_x)
    previous_step_size=abs(cur_x-new_x)
    print("iteration is:", num_of_iters)
    print("x is:", cur_x)
    num_of_iters=num_of_iters+1
    initial_x=cur_x

print("the minimum occurs at x =", cur_x)
    
```
